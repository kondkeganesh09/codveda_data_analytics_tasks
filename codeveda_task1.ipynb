```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("iris.csv")
data.head()
```

```python
# STEP 2: Basic Data Inspection
data.shape
```

```python
# Dataset information
data.info()
```

```python
# Check missing values
data.isnull().sum()
```

```python
# STEP 3: Data Cleaning
data.duplicated().sum()
```

```python
df = data.drop_duplicates()
```

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
df['species'].unique()
```

```python
# STEP 4: Summary Statistics
df.describe()
```

```python
# STEP 5: Univariate Analysis (Distributions)
# histogram
df.hist(figsize=(10,6))
plt.tight_layout()
plt.show()
```

```python
# outlier detection
plt.figure(figsize=(8,5))
sns.boxplot(data=df)
plt.title("Boxplot of Iris Features")
plt.show()
```

```python
# Bivariate Analysis
# Scatter Plot
plt.figure(figsize=(6,4))
sns.scatterplot(
    x='sepal_length',
    y='sepal_width',
    hue='species',
    data=df
)
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
# STEP 7: Correlation Analysis
correlation = df.drop('species', axis=1).corr()
correlation
```

```python
# Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# STEP 8: Basic Data Visualization
# Bar Plot (Count of species)
sns.countplot(x='species', data=df)
plt.title("Count of Each Iris Species")
plt.show()
```

```python
# Pairplot
sns.pairplot(df, hue='species')
plt.show()
```
